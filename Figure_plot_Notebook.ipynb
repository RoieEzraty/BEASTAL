```python
import matplotlib.patches as patches

def draw_vector_iris(ax, center=(0.35, 0.08), scale=1.0):
    """
    Draws a stylized vector Iris-like flower using patches on the given Axes.
    Coordinates are relative to 'axes fraction'.
    """
    from matplotlib.transforms import Affine2D

    # Transform to position and scale in axes coordinates
    trans = Affine2D().scale(scale).translate(*center) + ax.transAxes

    # Stylized petals (3 upward)
    petals = [
        patches.Ellipse((0, 0.2), 0.08, 0.18, angle=20, fill=False, lw=1.2, transform=trans),
        patches.Ellipse((0, 0.2), 0.08, 0.18, angle=-20, fill=False, lw=1.2, transform=trans),
        patches.Ellipse((0, 0.27), 0.1, 0.2, angle=0, fill=False, lw=1.2, transform=trans),
    ]

    # Side leaves
    leaves = [
        patches.Ellipse((0.05, -0.2), 0.04, 0.2, angle=-45, fill=False, lw=1.2, transform=trans),
        patches.Ellipse((-0.05, -0.2), 0.04, 0.2, angle=45, fill=False, lw=1.2, transform=trans),
    ]

    # Stem
    stem = patches.FancyArrow(0, -0.05, 0, -0.1, width=0.005, length_includes_head=True,
                              head_width=0.01, head_length=0.02, color='black', transform=trans)

    # Add all
    for p in petals + leaves:
        ax.add_patch(p)
    ax.add_patch(stem)
```

```python

# plot accuracy a.f.o time
fig, ax = plt.subplots()
draw_vector_iris(ax, center=(0.35, 0.08), scale=1.0)
```
